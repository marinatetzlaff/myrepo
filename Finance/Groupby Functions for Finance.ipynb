# Groupby Functions for Finance

Source: https://towardsdatascience.com/ultimate-pandas-guide-mastering-the-groupby-104306251739 

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
data = pd.read_csv('fundamentals.csv')
```

```python
data.head()
```

```python
#get total earnings per ticker
data.groupby('For Year').agg(sum)[['Earnings Per Share']]
#group by addresses the split stage
#agg addresses the “apply” stag
```

# Understanding the “split” step in Pandas

```python
grouped = data.groupby('For Year')
```

```python
grouped.groups
```

```python
grouped.get_group(2016)
```

```python
for name, group in grouped:
    print(name, group)
```

# Understanding the “agg” step in Pandas

```python
data.groupby('For Year').agg(sum)[['Earnings Per Share']]
#returns a data frame
```

```python
data.groupby('For Year').agg(sum)['Earnings Per Share']
#returns a Series object
```

# Advanced “groupby” concepts

```python
data.groupby('For Year').agg([sum, np.mean])[['Earnings Per Share','Total Revenue']]
```

#  'NamedAgg' object from Pandas

Using an object:

```python
data.groupby('For Year').agg(
   earnings_per_share=pd.NamedAgg(column='Earnings Per Share', aggfunc=sum),
   avg_total_revenue=pd.NamedAgg(column='Total Revenue', aggfunc=np.mean))
```

Dictionary to the agg function:

```python
data.groupby('For Year').agg( 
        {'Earnings Per Share': [sum],
        'Total Revenuee': [np.mean]}
)
```

Group by multiple columns

```python
data.groupby(['For Year', 'Ticker Symbol']).agg(sum)[['Earnings Per Share']]
```

Convert indices to columns using the "reset index" command

```python
g = data.groupby(['For Year', 'Ticker Symbol']).agg(sum) 
g['Earnings Per Share'].reset_index()
```
